```python
import cv2 as cv  # import thư viện opencv để xử lý ảnh
import matplotlib.pyplot as ThanhNguyen  # import matplotlib và đặt tên khác
# 2. Kiểm tra xem opencv đọc ảnh như thế nào? Hiển thị chiều của ảnh?
img = cv.imread('hinh1.jpg')  # đọc ảnh từ file
print("Kiểu dữ liệu:", type(img))  # in kiểu dữ liệu của ảnh
print("Shape (Height, Width, Channels):", img.shape)  # in kích thước ảnh

# 3. Show hình ảnh ra cho trực quan, sử dụng plt.imshow. 
ThanhNguyen.imshow(img)  # hiển thị ảnh
ThanhNguyen.title("Ảnh gốc hiển thị bằng Matplotlib")  # đặt tiêu đề
ThanhNguyen.axis("off")  # tắt trục tọa độ
ThanhNguyen.show()  # hiển thị ra màn hình

# Có nhận xét gì về bức ảnh? Có cách nào để hiển thị đúng hay không? 
# Nhận xét:
# Ảnh hiển thị sai màu (bị xanh/đỏ lệch).
# Nguyên nhân: OpenCV dùng BGR, Matplotlib dùng RGB.
# Cách 1, đảo chiều matrix; 
img_rgb1 = img[:, :, ::-1]  # đảo thứ tự kênh màu BGR → RGB
ThanhNguyen.imshow(img_rgb1)  # hiển thị ảnh sau khi đảo kênh
ThanhNguyen.title("Hiển thị đúng màu - Cách 1")  # đặt tiêu đề
ThanhNguyen.axis("off")  # tắt trục tọa độ
ThanhNguyen.show()  # hiển thị kết quả

# cách 2, convert các chế độ màu. 
img_rgb2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # chuyển không gian màu BGR sang RGB
ThanhNguyen.imshow(img_rgb2)  # hiển thị ảnh sau khi convert
ThanhNguyen.title("Hiển thị đúng màu - Cách 2")  # đặt tiêu đề
ThanhNguyen.axis("off")  # tắt trục tọa độ
ThanhNguyen.show()  # hiển thị kết quả

# 4. Nếu muốn phóng to ảnh thì chúng ta làm gì? 
img_big = cv.resize(img, None, fx=3, fy=3)  # phóng to ảnh theo tỉ lệ
ThanhNguyen.figure(figsize=(10, 10)) # tạo khung ảnh to 
ThanhNguyen.imshow(cv.cvtColor(img_big, cv.COLOR_BGR2RGB))  # hiển thị ảnh phóng to
ThanhNguyen.title("Ảnh phóng to")  # đặt tiêu đề
ThanhNguyen.axis("off")  # tắt trục tọa độ
ThanhNguyen.show()  # hiển thị kết quả

# 5. Nếu muốn crop ảnh thì chúng ta làm gì? 
upper = 10 # chỉ số hàng trên để cắt ảnh
lower = 400 # chỉ số hàng dưới để cắt ảnh
crop_top = img[upper: lower,:,:] # cắt vùng ảnh theo chiều dọc
left = 200 # chỉ số cột bên trái để cắt ảnh
right = 900 # chỉ số cột bên phải để cắt ảnh
crop_horizontal = crop_top[:,left:right,:] # cắt tiếp theo chiều ngang
ThanhNguyen.imshow(cv.cvtColor(crop_horizontal, cv.COLOR_BGR2RGB)) # hiển thị ảnh đã cắt
ThanhNguyen.title("Crop ảnh") # đặt tiêu đề
ThanhNguyen.axis("off") # tắt trục tọa độ
ThanhNguyen.show() # hiển thị kết quả

# 6. Ghép dọc và ghép ngang 2 bức ảnh.
img2 = cv.imread("hinh1.jpg") # đọc ảnh thứ hai
img2 = cv.resize(img2, (img.shape[1], img.shape[0])) # resize ảnh thứ hai cho cùng kích thước

v_concat = cv.vconcat([img, img2]) # ghép ảnh theo chiều dọc
ThanhNguyen.imshow(cv.cvtColor(v_concat, cv.COLOR_BGR2RGB)) # hiển thị ảnh ghép dọc
ThanhNguyen.title("Ghép dọc") # đặt tiêu đề
ThanhNguyen.axis("off") # tắt trục tọa độ
ThanhNguyen.show() # hiển thị kết quả

h_concat = cv.hconcat([img, img2]) # ghép ảnh theo chiều ngang
ThanhNguyen.imshow(cv.cvtColor(h_concat, cv.COLOR_BGR2RGB)) # hiển thị ảnh ghép ngang
ThanhNguyen.title("Ghép ngang") # đặt tiêu đề
ThanhNguyen.axis("off") # tắt trục tọa độ
ThanhNguyen.show() # hiển thị kết quả
```
